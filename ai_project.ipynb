* In the function I remove punctuation, use lemmatization and remove stop words
* Although since it was really slow, I used nltk for text cleaning.

```python
import spacy
import el_core_news_sm

def extract_clean_body(speech):
        
    nlp = spacy.load("el_core_news_sm")    
    text = speech
    my_doc = nlp(text)
    
    #remove punctuation
    text_without_punt = [token for token in my_doc if not token.is_punct]
    
    
    # Create list of word tokens and lemma
    token_list = []
    lemma_list = []
    for token in text_without_punt:
        token_list.append(token.text)
        lemma_list.append(token.lemma_)
        
    from spacy.lang.en.stop_words import STOP_WORDS

    # Create list of word tokens after removing stopwords
    filtered_sentence =[] 
    joined_sentence =[]
    for word in token_list:
        lexeme = nlp.vocab[word]
        if lexeme.is_stop == False:
            filtered_sentence.append(word) 
        # Join words.
        joined_sentence = ' '.join(filtered_sentence)
        
    return joined_sentence
    
```

* Create pipelines for the different algorithms and for tuning hyperparameters.
* Regarding n gram range: I used different approaches in a very small sample only with MultinomialNB since this parameter slowed down dramatically lightGBM. (1,3) n grams was the optimal for MultinomialNB for a very small sample, but when I used it in GridSearch it slowed it down dramatically so I used (1,1) which is faster.
* Regarding max_df: I used it on a sample of the dataset but it didn't seem to make any significant difference.
* Regarding max features parameter: I started working on a very small sample of only 5.000 speeches and the classifiers would take hours. I solved this problem by adding max_features on countvectorizer and tfidfvectorizer.I started by setting it to 10.000 words. I continued by doing grid search on that small sample of speeches and found out that max feautures between 1000 and 2000 words was the optimal (and more specifically around 1500 words)! So I use 1300, 1500 words as max feautures. The optimal was around 1500 for the larger sample (data from 2010 and on) too. For example for LightGBM when trying between 1300 and 1500, 1300 was the best choice.
* I have used one tree based algorithm, LightGBM. In order to prevent overfitting and speed up the classifier, I used a small number of depth and leaves. Tried to use a larger number for tree depth, more specyfically 10 but it slowed down significantly, so I stick to 7 for depth. Since LightGBM needs 20 minutes to run, I can't do excessive hyperparameter tuning. I will just do cross validation on the best parameters that I found when testing it in a smaller sample. 
* I also tried to use XGBClassifier but it was very slow
* Lastly I use an algorithm based on NB thorem, MultinomialNB which is observed as a good text classifier. I only did some parameter tuning regarding alpha

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.svm import SVC 
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.dummy import DummyClassifier

countv_MNB_pipe = Pipeline([
 ('cvec', CountVectorizer()),
 ('mnb', MultinomialNB()),
])


tfv_MNB_pipe = Pipeline([
 ('tvec', TfidfVectorizer()),
 ('mnb', MultinomialNB()),
])


tfv_xgbc_pipe = Pipeline([
 ('tvec', TfidfVectorizer()),
 ('xgb', XGBClassifier()),
])


cv_lgbmc_pipe = Pipeline([
 ('cvec', CountVectorizer()),
 ('lgbm', LGBMClassifier())
])


tfv_lgbmc_pipe = Pipeline([
 ('tvec', TfidfVectorizer()),
 ('lgbm', LGBMClassifier())
])

cv_dummy_pipe = Pipeline([
    ('cvec', CountVectorizer()),
    ('dummy', DummyClassifier())
])


# Setting params for Count Vectorizer and MultinomialNB
countv_MNB_params = {
    'cvec__min_df' : [0.01],
    'cvec__max_features' : [1300, 1500],
    'mnb__alpha' : [0.1, 0.001]  
}

# Setting params for tfidf Vectorizer and MultinomialNB
tfv_MNB_params = {
    'tvec__min_df' : [0.01],
    'tvec__max_features' : [1300, 1500],
    'mnb__alpha' : [0.1 , 0.001]
}


# Setting params for tfidf Vectorizer and XGBClassifier
tfv_xgbc_params = {
    'tvec__min_df' : [0.01],
    'tvec__max_features' : [1300],
    'xgb__max_depth' : [5],
    'xgb__learning_rate' : [0.1]
}

# Setting params for Count Vectorizer and LightGBM
cv_lgbmc_params = {
    'cvec__min_df' : [0.01],
    'cvec__max_features' : [1300, 1500],
    'lgbm__max_depth' : [5, 7],
    'lgbm__num_leaves' :[12, 20],
    'lgbm__min_gain_to_split' : [0.01, 0.001]
}


# Setting params for Tf-idf Vectorizer and LightGBM
tfv_lgbmc_params = {
    'tvec__min_df' : [0.01],
    'tvec__max_features' : [1300],
    'lgbm__max_depth' : [5, 7],
    'lgbm__num_leaves' :[12, 20],
    'lgbm__min_gain_to_split' : [0.01, 0.001]
}

# Setting params for dummy classifier
cv_dummy_params = {
    'cvec__min_df' : [0.01],
    'cvec__max_features' : [1300]
}
```

* Use nested cross validation with grid search for hyperameter tuning

* define a function for grid search and cross validation

```python
# manual cross-validation 
from sklearn.model_selection import GridSearchCV
from numpy import mean
from numpy import std
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
from sklearn import metrics
from sklearn.metrics import classification_report

def cross_val_grid_search(pipeline, params):
    # dataset
    X = g_p_for_classification
    y = g_p_for_classification
    # configure the cross-validation procedure
    cv_outer = KFold(n_splits=3, shuffle=True, random_state=1)
    # enumerate splits
    outer_results = list()
    for train_ix, test_ix in cv_outer.split(X):


        # split data
        X_train, X_test = X.iloc[train_ix, X.columns.get_loc('speech')], X.iloc[test_ix, X.columns.get_loc('speech')]
        y_train, y_test = y.iloc[train_ix,y.columns.get_loc('political_party')], y.iloc[test_ix,y.columns.get_loc('political_party')]


        # define search
        search = GridSearchCV(pipeline, param_grid=params, scoring='accuracy', verbose =1, n_jobs = -1)

        # execute search
        result = search.fit(X_train, y_train)

        # get the best performing model fit on the whole training set
        best_model = result.best_estimator_
        print('The best model is:')
        print(best_model)
        print()

        # evaluate model on the hold out dataset
        yhat = best_model.predict(X_test)

        # evaluate the model
        acc = accuracy_score(y_test, yhat)

        # store the result
        outer_results.append(acc)
        
        #classification report
        print("Detailed classification report:")
        print(classification_report(y_test, yhat))
        print()

        # report progress
        print('Accuracy, Best Score, Best Parameters:')
        print('>acc=%.3f, est=%.3f, cfg=%s' % (acc, result.best_score_, result.best_params_))
        print()
        
        #confusion matrix
        print('Confusion Matrix:')
        print(metrics.confusion_matrix(y_test, yhat))
        print()

    # summarize the estimated performance of the model
    print('Accuracy: %.3f (%.3f)' % (mean(outer_results), std(outer_results)))
```

* Create a function to pass training and testing data, as well as the column-feature (which is speeches, but I also added parliament sitting in one model). This function encodes the words to numbers, and makes them ready to feed in the embedding layer. We also one hot ecode the target variable (political party).
* This function will be used again later, when I will use more than one feature to feed the neural network.

```python
def encode_words_to_numbers(X_train, X_test, y_train, y_test, column):
  tokenizer = Tokenizer(num_words=vocab_size, oov_token='<OOV>')
  tokenizer.fit_on_texts(X_train[column])
  word_index = tokenizer.word_index
  print('Found %s unique tokens.' % len(word_index))
  dict(list(word_index.items())[0:10])

  train_seq = tokenizer.texts_to_sequences(X_train[column])
  train_padded = pad_sequences(train_seq, maxlen=max_length, padding=padding_type, truncating=trunc_type)

  validation_seq = tokenizer.texts_to_sequences(X_test[column])
  validation_padded = pad_sequences(validation_seq, maxlen=max_length, padding=padding_type, truncating=trunc_type)
  print('Shape of data tensor:', train_padded.shape)
  print('Shape of data tensor:', validation_padded.shape)

  encode = OneHotEncoder()

  training_labels = encode.fit_transform(y_train)
  validation_labels = encode.transform(y_test)

  # The labels must be converted to arrays
  # Convert the labels to arrays
  training_labels = training_labels.toarray()
  validation_labels = validation_labels.toarray()

  return train_padded, training_labels, validation_padded, validation_labels
```

```python
from tensorflow.keras import layers
from kerastuner.tuners import RandomSearch
from tensorflow import keras


def create_model(hyperparam):

  GRU_NEURONS = hyperparam.Choice('neurons',[32,64,128])
  DROPOUT_RATE = hyperparam.Choice('dropout',[0.1,0.2])
  EMBEDDING_DIM = hyperparam.Choice('embedding',[100, 200, 300])
  DENSE_ = hyperparam.Choice('dense',[32, 64, 128])


  model = keras.Sequential()

  model.add(Embedding(1500, EMBEDDING_DIM, input_length=train_padded.shape[1]))
  model.add(GRU(GRU_NEURONS, activation='tanh'))
  model.add(Flatten())
  model.add(Dropout(rate=DROPOUT_RATE))
  model.add(Dense(DENSE_, activation='tanh'))
  model.add(Dense(17, activation='softmax'))
  model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

  return model
```
